# 3.Check whether all the zones generate the same amount of sales.

So here we create a hypothesis

H0: All zones generate same amount of sales

Ha: All zones generate different amount of sales 

```python
f_value , p_value = scp.f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'],data['Zone - D'])
print('The f value is:',f_value)
print('The p value is:',p_value)
```

```python
if(p_value<0.05):
   print('Reject H0:All zones generate different amount of sales ')
else:
   print('Fail to Reject H0: All zones generate same amount of sales')
```
